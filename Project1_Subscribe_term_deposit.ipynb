From the correlation matrix, we notice that (euribor3m and nr.employed= 0.94), (euribor3m and emp.var.rate= 0.97), (emp.var.rate and nr.employed= 0.90). This indicate that is has multicollinearity among these feaure, so we need to drop some feature to avoid redundancy. Lets decide which to drop first. Alternatively, if we choose to drop emp.var.rate, the highest correlation is with cons.price.idx(0.47). 

So we choose to drop emp.var.rate because it has slightly lower average correlations with other features.

# Data Preprocessing

->Impute the 'unknown' category

->Target variable Encoding (Map y to become 0,no and 1,yes)

->Encode the categorical variable to numeric value  

->Handle Class Imbalance using SMOTE

Check which feature is categorical

```python
# Find the categorical variables
objects = ['object','category', 'bool']
cat_features = df2.select_dtypes(include=objects).columns.tolist()

# Category feature
cat_features

for col in cat_features:
    print(f"--- {col} ---")
    print(df2[col].value_counts())
    print("\n")
```

From the result, there is 'unknown' category in some feature. We need to check thier proportion is big or small. If <20% we can use mode to impute.

Feature poutcome have 3352 nonexistent values. It most likely means the client was not previously contacted, so i think it is meaningful, no need to drop it. 

Lets see thier weightage of the "unknown" in category feature

```python
total_rows = len(df2)

for col in df2.select_dtypes(include='object').columns:
    if 'unknown' in df2[col].unique():
        count = df2[col].value_counts().get('unknown', 0)
        percent = round((count / total_rows) * 100, 2)
        print(f"{col}: has {count} 'unknown', Missing propotion is: {percent}%")
```

The results shows, our unknown proportion is less than 20 percent. We can just use mode imputation for these categorical data. We assumes that the missing values are likely to belong to the most common category. 

```python
# Now we impute unknown using mode
for col in cat_features:
    if col != 'y':  # impute the category feature except the target variable.
        df2[col] = df2[col].replace('unknown', df2[col].mode()[0])

# See the newest dataframe still got unknown or not
for col in cat_features:
    print(f"--- {col} ---")
    print(df2[col].value_counts())
    print("\n")
```

```python
df2.info()
```

Target variable Encoding

```python
# Define features and target
x = df2.drop('y', axis=1)
y = df2['y'].map({'no':0, 'yes':1}) 

y
```

Now we use label encode to transform the category feature into binary.                


```python
from sklearn.preprocessing import LabelEncoder

current_cat_features = [col for col in cat_features if col in x]
print(current_cat_features)

# Label encode all categorical columns in x
label_encoder = LabelEncoder()
x_encoded = x.copy()
for col in current_cat_features:
    x_encoded[col] = label_encoder.fit_transform(x_encoded[col])

x_encoded.head()
```

Split Data before handling class imbalance

```python
from sklearn.model_selection import train_test_split

# Split data before any data preprocessing
x_train_encoded, x_test_encoded, y_train_encoded, y_test_encoded = train_test_split(x_encoded, y, test_size=0.2, stratify=y, random_state=42)

print("x_train shape:", x_train_encoded.shape)
print("x_test shape:", x_test_encoded.shape)
print("y_train value counts:\n", y_train_encoded.value_counts())
print("y_test value counts:\n", y_test_encoded.value_counts())
```

Handling class imbalances by applying SMOTE to the train dataset.

```python
# Apply SMOTE to handle class imbalance
from collections import Counter
from imblearn.over_sampling import SMOTE

print("Before ", Counter(y_train_encoded))

#Apply SMOTE on train set
smote = SMOTE(random_state=42)
x_smote, y_smote = smote.fit_resample(x_train_encoded, y_train_encoded)

print("After ", Counter(y_smote))
```

Before SMOTE is 2934 no, 361 yes. After SMOTE both yes and no = 2934

# Model Development

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

k_fold_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

Model 1 : Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logmodel = LogisticRegression(random_state=42) 
logmodel.fit(x_smote, y_smote)  # Trains the Logistic Regression model on the entire SMOTE-balanced training set.

# Performs cross-validation (CV) on the training set (x_smote, y_smote).
Logcv_scores = cross_val_score(logmodel, x_smote, y_smote, cv=k_fold_cv, n_jobs=-1, scoring='accuracy')
mean_cv = Logcv_scores.mean()
std_cv = Logcv_scores.std()
print(f"CV Accuracy: {mean_cv:.4f} (Â±{std_cv:.4f})")

# Model performance on testset
logpred = logmodel.predict(x_test_encoded)
log_test_acc = accuracy_score(y_test_encoded, logpred)
print(f"Test Accuracy: {log_test_acc:.4f}")
```

Model 2: Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(random_state=42)
dtree.fit(x_smote, y_smote) 

# Performs cross-validation (CV) on the training set (x_smote, y_smote).
Dtreecv_scores = cross_val_score(dtree, x_smote, y_smote, cv=k_fold_cv, n_jobs=-1, scoring='accuracy')
mean_cv = Dtreecv_scores.mean()
std_cv = Dtreecv_scores.std()
print(f"CV Accuracy: {mean_cv:.4f} (Â±{std_cv:.4f})")

# Model performance on testset
Dtreepred = dtree.predict(x_test_encoded)
Dtree_test_acc = accuracy_score(y_test_encoded, Dtreepred)
print(f"Test Accuracy: {Dtree_test_acc:.4f}")
```

Model 3: Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=42)
rfc.fit(x_smote, y_smote)

# Performs cross-validation (CV) on the training set (x_smote, y_smote).
rfccv_scores = cross_val_score(rfc, x_smote, y_smote, cv=k_fold_cv, n_jobs=-1, scoring='accuracy')
mean_cv = rfccv_scores.mean()
std_cv = rfccv_scores.std()
print(f"CV Accuracy: {mean_cv:.4f} (Â±{std_cv:.4f})")

# Model performance on testset
rfcpred = rfc.predict(x_test_encoded)
rfc_test_acc = accuracy_score(y_test_encoded, rfcpred)
print(f"Test Accuracy: {rfc_test_acc:.4f}")
```

Model 4: XGBoost

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(random_state=42)
xgb.fit(x_smote, y_smote)

# Performs cross-validation (CV) on the training set (x_smote, y_smote).
xgbcv_scores = cross_val_score(xgb, x_smote, y_smote, cv=k_fold_cv, n_jobs=-1, scoring='accuracy')
mean_cv = xgbcv_scores.mean()
std_cv = xgbcv_scores.std()
print(f"CV Accuracy: {mean_cv:.4f} (Â±{std_cv:.4f})")

# Model performance on testset
xgbpred = xgb.predict(x_test_encoded)
xgb_test_acc = accuracy_score(y_test_encoded, xgbpred)
print(f"Test Accuracy: {xgb_test_acc:.4f}")
```

Model 5: Light GBM

```python
from lightgbm import LGBMClassifier
lgbm = LGBMClassifier(random_state=42)
lgbm.fit(x_smote, y_smote)

# Performs cross-validation (CV) on the training set (x_smote, y_smote).
lgbmcv_scores = cross_val_score(lgbm, x_smote, y_smote, cv=k_fold_cv, n_jobs=-1, scoring='accuracy')
mean_cv = lgbmcv_scores.mean()
std_cv = lgbmcv_scores.std()
print(f"CV Accuracy: {mean_cv:.4f} (Â±{std_cv:.4f})")

# Model performance on testset
lgbmpred = lgbm.predict(x_test_encoded)
lgbm_test_acc = accuracy_score(y_test_encoded, lgbmpred)
print(f"Test Accuracy: {lgbm_test_acc:.4f}")
```

Model 6: Cat Boost

```python
from catboost import CatBoostClassifier

catboost = CatBoostClassifier(random_state=42)
catboost.fit(x_smote, y_smote)

# Performs cross-validation (CV) on the training set (x_smote, y_smote).
catboostcv_scores = cross_val_score(catboost, x_smote, y_smote, cv=k_fold_cv, n_jobs=-1, scoring='accuracy')
mean_cv = catboostcv_scores.mean()
std_cv = catboostcv_scores.std()
print(f"CV Accuracy: {mean_cv:.4f} (Â±{std_cv:.4f})")

# Model performance on testset
catboostpred = catboost.predict(x_test_encoded)
catboost_test_acc = accuracy_score(y_test_encoded, catboostpred)
print(f"Test Accuracy: {catboost_test_acc:.4f}")
```

# Model Selection

Among 6 model, CatBoost is selected as the best model because it achive the highest test accuracy on test set with 0.8641 and also a good cross-validation mean (0.9095) with reasonable standard deviation (0.0095).                   
Lets do a hyperparameter tunning on catboost model

```python
# Hyperparameter Tunning for Catboost Model
from sklearn.model_selection import GridSearchCV

# Use the already SMOTE-balanced data
new_catboost = CatBoostClassifier(scale_pos_weight=8,iterations=1000,early_stopping_rounds=50,eval_metric='AUC',verbose=False,random_state=42)

param_grid = {
    'depth': [5,8,10],
    'learning_rate': [0.03,0.05, 0.1],
    'l2_leaf_reg': [3,5,7]
}

# Grid search with 5-fold cross-validation, scoring by accuracy
grid_search = GridSearchCV(new_catboost, param_grid, cv=k_fold_cv, scoring='accuracy',n_jobs=-1)
grid_search.fit(x_smote, y_smote)
best_params = grid_search.best_params_
catboostmodel_with_tune = grid_search.best_estimator_

print("Best parameters found:", best_params)
print("Best accuracy score:", grid_search.best_score_)

# Model performance on testset
catboostmodel_with_tune_pred = catboostmodel_with_tune.predict(x_test_encoded)
catboost1_test_acc = accuracy_score(y_test_encoded, catboostmodel_with_tune_pred)
print(f"Test Accuracy: {catboost1_test_acc:.4f}")
```

From the result, we notice that the tuned model has a slightly improvement in train set(0.9095->0.9158), but on the test set the tunned model is perform worse than the default model(0.8641->0.8277). The model is overfitting to the SMOTE training data and maybe label encoding distort the category relationship. Lets try remove the SMOTE and no encoding on training set, Since CatBoost itself has built-in handling for categorical features and class imbalance (via `scale_pos_weight`).
